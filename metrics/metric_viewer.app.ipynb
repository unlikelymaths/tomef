This script loads the metric data file and creates a table showing all metrics of the specified category.

```python
from __init__ import init_vars
init_vars(vars())

import numpy as np

import data
import config
from base import nbprint
from widgetbase import nbbox

# clustering, classification, coherence
category = 'coherence'
# None, nmi, nmi, umass_wiki, svm_5fold_micro_f1
sort_by = 'umass_wiki'
# filter lines based function. True == Keep line
token_filter = lambda entry: entry['info']['num_tokens'] == 10
data_filter = lambda entry: entry['info']['data_name'] == 'classic4'
entry_filter = []

# metric_fct applied to each metric
def average(m):
    val = sum(m) / len(m)
    if np.isnan(val):
        return -np.inf
    return val
def iden(m):
    return m
metric_fct = average

# Load data. Each entry in metric_data is one row of the table
metric_data = []
for entry in data.load_metric_data(category):
    keep = True
    for f in entry_filter:
        keep = keep and f(entry)
    if keep:
        metric_data.append(entry)

# column headers of the table
info_keys = []
metric_keys = []
for entry in metric_data:
    for key in entry['info']:
        if key not in info_keys:
            info_keys.append(key)   
    for key in entry['data']:
        if key not in metric_keys:
            metric_keys.append(key)        

# Get line data
lines = []
for entry in metric_data:
    line = []
    for key in info_keys:
        line.append(entry['info'].get(key, '-'))
    for key in metric_keys:
        metric = entry['data'].get(key, '-')
        if metric_fct:
            metric = metric_fct(metric)
        line.append(metric)
    lines.append(line)

# Sort
if sort_by is not None:
    idx = len(info_keys) + metric_keys.index(sort_by)
    lines.sort(key=lambda e: e[idx], reverse=True)

# Format using Markdown
nbbox()
nbprint('| ' + ' | '.join(info_keys + metric_keys) + ' |',prefix = False)
nbprint('|---'*(len(info_keys) + len(metric_keys)) + '|',prefix = False)
for line in lines:
    nbprint('| ' + ' | '.join([str(e) for e in line]) + ' |',prefix = False)
```
